### Regression lineaire Simple 

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = {
    'Age':[22,25,30,35,40,45,50],
    'Salaire (MAD)':[4000,5000,6000,7000,8000,9500,11000]
}
df = pd.DataFrame(data)

X = df[['Age']] # Variable explicative (X doit etre un tableau)
Y = df['Salaire (MAD)']

model = LinearRegression()
model.fit(X,Y)

print(f"Equation: y = {model.coef_[0]:.2f} * x + {model.intercept_:.2f}")

# Étape 6 : Visualisation
plt.scatter(X, Y, color='blue', label='Données réelles')
plt.plot(X, model.predict(X), color='red', label='Droite de régression')

plt.xlabel("Âge")
plt.ylabel("Salaire (MAD)")
plt.title("Régression Linéaire : Âge vs Salaire")

plt.legend()
plt.grid()
plt.show()
```

### Regression lineaire Multiple

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Données simulées
data = {
    'Âge': [25, 30, 35, 40, 45],
    'Expérience': [1, 3, 5, 7, 10],
    'Niveau_Etude': [0, 1, 1, 2, 2],
    'Salaire': [4000, 5000, 6000, 7000, 8000]
}

df = pd.DataFrame(data)

# Variables explicatives (X) et variable cible (y)
X = df[['Âge', 'Expérience', 'Niveau_Etude']]
y = df['Salaire']

# Modèle de régression multiple
model = LinearRegression()
model.fit(X, y)

# Coefficients
print("Coef :", model.coef_)
print("Intercept :", model.intercept_)

# Prédictions
y_pred = model.predict(X)

# Scatter plot : réel vs prédiction
plt.scatter(y, y_pred, color='blue')

# Ligne idéale (y = y)
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red')

plt.xlabel("Valeurs réelles (Salaire)")
plt.ylabel("Prédictions du modèle")
plt.title("Régression multiple : Réel vs Prédit")

plt.grid()
plt.show()
```
